```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as pt
l1=[x for x in range(0,20)]
l2=[x*x for x in range(0,20)]
data=pd.DataFrame({'x':l1,'x_square':l2},columns=['x','x_square'])
print(data)
```

```python
pt.plot(data['x'],data['x_square'])
```

```python
X=data.iloc[0:18, :1].values
Y=data.iloc[0:18:, 1:2].values
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.33,random_state=0)
print(X_train)
print(Y_train)
```

```python
from sklearn.linear_model import LinearRegression
#object
reg=LinearRegression()
#fitting the deman line for curve
reg.fit(X_train,Y_train)
# prediction form the test labels
y_pred=reg.predict(X_train)
print(X_train,y_pred)
```

```python
pt.figure(figsize=(20,10))
pt.title('linear regression train data',fontsize=24)
pt.scatter(X_train,Y_train,color='red')
pt.plot(X_train,reg.predict(X_train),color='blue',label='demand line')
pt.xlabel('x')
pt.legend(loc='best',prop={'size': 26})
pt.grid(True)
```

```python
#visualising test
pt.figure(figsize=(20,10))
pt.title('linear regression test plot',fontsize=24)
pt.scatter(X_test,Y_test,color='red')
pt.plot(X_train,reg.predict(X_train),color='blue',label='demand line')
pt.legend(loc='best',prop={'size': 26})
pt.grid(which='major',color='blue')
pt.minorticks_on()
pt.grid(which='minor',color='green',linestyle='--')
pt.xlabel('x')
pt.grid(True)
```
