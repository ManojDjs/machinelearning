```python
import pandas as pd
import numpy as np
 
#Create a DataFrame
df1 = {
     'Name':['George','Andrea','micheal','maggie','Ravi','Xien','Jalpa'],
   'University_Rank':[6,47,21,74,32,77,8]}
 
df1 = pd.DataFrame(df1,columns=['Name','University_Rank'])
print(df1)
```

```python
df1['log_value'] = np.log(df1['University_Rank'])
print(df1)
```

```python
import matplotlib.pyplot as pl
import seaborn as sns
p1=sns.distplot(df1['log_value'])
print(p1)
```

```python
p2=sns.distplot(df1['University_Rank'])
print(p2)
```

```python
df1['exp_value'] = np.exp(df1['University_Rank'])
```

```python
p2=sns.distplot(df1['exp_value'])
print(p2)
```
